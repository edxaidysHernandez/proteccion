```python
import pandas as pd
import json

# ── Carga ─────────────────────────────────────────────────────────────────────
items       = pd.read_csv('order_items_dataset.csv')
products    = pd.read_csv('products_dataset.csv')
translation = pd.read_csv('product_category_name_translation.csv')
translation.columns = translation.columns.str.replace('\ufeff', '')

# ── Join ──────────────────────────────────────────────────────────────────────
df = (
    items
    .merge(products[['product_id', 'product_category_name']], on='product_id')
    .merge(translation, on='product_category_name', how='left')
)
df['category'] = df['product_category_name_english'].fillna(df['product_category_name'])

# ── Precios reales por categoría ──────────────────────────────────────────────
def precios(cat):
    sub = df[df['category'] == cat]['price']
    return round(sub.mean(), 2), round(sub.median(), 2)

# ── Base de conocimiento ──────────────────────────────────────────────────────
knowledge_base = {"base_conocimiento": []}

categorias = [
    {
        "id": "CAT-001",
        "categoria": "health_beauty",
        "nombre_display": "Salud y Belleza",
        "descripcion": "Descubre nuestra selección premium de productos para el cuidado personal. Desde cremas hidratantes y suplementos vitamínicos hasta kits de maquillaje profesional, esta categoría reúne todo lo que necesitas para verte y sentirte mejor cada día.",
        "tags": ["cuidado personal", "bienestar", "belleza", "salud"]
    },
    {
        "id": "CAT-002",
        "categoria": "watches_gifts",
        "nombre_display": "Relojes y Regalos",
        "descripcion": "La categoría perfecta para momentos especiales. Relojes elegantes, accesorios exclusivos y artículos de regalo cuidadosamente seleccionados para cada ocasión. Con el ticket promedio más alto del catálogo, estos productos atraen a clientes que buscan calidad y distinción.",
        "tags": ["regalos", "relojes", "accesorios", "premium"]
    },
    {
        "id": "CAT-003",
        "categoria": "bed_bath_table",
        "nombre_display": "Cama, Baño y Mesa",
        "descripcion": "Transforma tu hogar con nuestra colección más vendida. Ropa de cama de alta calidad, toallas suaves, vajillas elegantes y accesorios para el baño que combinan funcionalidad con estilo. La categoría líder en volumen de ventas.",
        "tags": ["hogar", "cama", "baño", "decoración"]
    }
]

for cat in categorias:
    promedio, mediano = precios(cat['categoria'])
    cat['precio_promedio_brl'] = promedio
    cat['precio_mediano_brl']  = mediano
    knowledge_base['base_conocimiento'].append(cat)

with open('knowledge_base.json', 'w', encoding='utf-8') as f:
    json.dump(knowledge_base, f, ensure_ascii=False, indent=2)

print(json.dumps(knowledge_base, ensure_ascii=False, indent=2))
```
